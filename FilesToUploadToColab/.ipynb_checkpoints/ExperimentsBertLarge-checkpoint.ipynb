**Define Experiment Function**

```python
def predict(sent, masked, ids, tail_masked_ids, bert):
    logprob = 0
    masked = deepcopy(masked)
    for m_id in tail_masked_ids:
        #get correct token id
        masked_token_id = sent[m_id]
        # make prediction
        pred = bert(masked.reshape(1,-1),ids.reshape(1,-1)).log_softmax(2)
        logprob = logprob + pred[0, m_id, masked_token_id]
        masked[m_id] =  masked_token_id
    return logprob

def bert_predictions(sentences_mask_tail, sentences_mask_head, sentences_mask_both, bert):
    data = []
    for idx, ((sent, masked_tail, ids, label), (_, masked_head, _, _), (_, masked_both, _, _)) \
            in enumerate(zip(sentences_mask_tail, sentences_mask_head, sentences_mask_both)):
        tail_masked_ids = [idx for idx, token in enumerate(masked_tail) if token == 103]
        head_masked_ids = [idx for idx, token in enumerate(masked_head) if token == 103]
        
        # conditional
        logprob_tail_conditional = predict(sent, masked_tail, ids, tail_masked_ids, bert)
        logprob_head_conditional = predict(sent, masked_head, ids, head_masked_ids, bert)
        # marginal
        logprob_tail_marginal = predict(sent, masked_both, ids, tail_masked_ids, bert)
        logprob_head_marginal = predict(sent, masked_both, ids, head_masked_ids, bert)

        NLL = -logprob_tail_conditional/len(tail_masked_ids)

        mutual_inf = logprob_tail_conditional - logprob_tail_marginal
        mutual_inf += logprob_head_conditional - logprob_head_marginal
        mutual_inf /= 2.
        try:
            print(idx, (NLL.item(), mutual_inf.item(), label, sentences_mask_tail.id_to_text(sent)))
            data.append((NLL.item(), logprob_tail_conditional.item(), logprob_tail_marginal.item(),
                         logprob_head_conditional.item(), logprob_head_marginal.item(),
                         mutual_inf.item(), label, sentences_mask_tail.id_to_text(sent)))
        except AttributeError:
            print(idx, (NLL, mutual_inf, label, sentences_mask_tail.id_to_text(sent)))
            data.append((NLL,  logprob_tail_conditional.item(), logprob_tail_marginal.item(),
                         logprob_head_conditional.item(), logprob_head_marginal.item(),
                         mutual_inf.item(), label, sentences_mask_tail.id_to_text(sent)))
    # prep data
    df = pd.DataFrame(data, columns = ('nll','tail_conditional','tail_marginal',
                                       'head_conditional','head_marginal','mut_inf','label','sent')) 
    return df
  
def experiment(Template, dev_data_path, test_data_path, bert, template_name):
    # get dev sentences
    sentences_mask_tail = Template(dev_data_path, False, True, 'cuda')
    sentences_mask_head = Template(dev_data_path, True, False, 'cuda')
    sentences_mask_both = Template(dev_data_path, True, True, 'cuda')
    
    df_train = bert_predictions(sentences_mask_tail, sentences_mask_head, sentences_mask_both, bert)
    df_train.to_csv(f'../{template_name}_train.csv')
    
    # prep data
    X_train = df_train[['mut_inf']].values
    y_train = df_train['label'].values
    
    #fit log reg
    mms = MinMaxScaler()
    X_train = mms.fit_transform(X_train)
    model = LogisticRegression(C=1000, random_state=0, solver='lbfgs',max_iter = 1000).fit(X_train, y_train)
    
    # get test sentences
    sentences_mask_tail = Template(test_data_path, False, True, 'cuda')
    sentences_mask_head = Template(test_data_path, True, False, 'cuda')
    sentences_mask_both = Template(test_data_path, True, True, 'cuda')
    
    df_test = bert_predictions(sentences_mask_tail, sentences_mask_head, sentences_mask_both, bert)

    X_test = df_test[['mut_inf']].values
    
    df_test['pred'] = model.predict(X_test)
    df_test['prob'] = model.predict_proba(X_test)[:,1]
    df_test.to_csv(f'../{template_name}_test.csv')

    get_acc = lambda df: sum(df.pred == df.label)/len(df)
    acc = get_acc(df_test)
    top_k = 100
    df_test_certain = df_test.sort_values('prob', ascending=False)[:100]
    acc_certain = get_acc(df_test_certain)
    return acc, acc_certain, model, df_test, df_train
  
def mining(Template, unknown_tuples_path, bert, logistic_reg_model, newfile_name):
    # get dev sentences
    sentences_conditional = Template(unknown_tuples_path, False, True, 'cuda')
    sentences_joint = Template(unknown_tuples_path, True, True, 'cuda')
    
    df_mining = bert_predictions(sentences_conditional, sentences_joint, bert)
    
    X_mining = df_mining[['nll','mut_inf']].values
    
    df_mining['prob'] = logistic_reg_model.predict_proba(X_mining)[:,1]
    df_mining_top100 = df_mining.sort_values('prob', ascending=False)[:100]
    df_mining_top100.to_csv(newfile_name)
    
```
